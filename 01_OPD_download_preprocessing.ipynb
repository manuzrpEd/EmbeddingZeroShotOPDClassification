```python
from preprocessing import (download_socrata_full_safe, filter_relevant_datasets, get_opd_catalog, 
    load_crime_data_for_nlp, summarize_downloads)
```

```python
catalog = get_opd_catalog()
catalog
```

```python
relevant_datasets = filter_relevant_datasets(catalog)
relevant_datasets
```

```python
downloaded = []
failed = []
for idx, row in relevant_datasets.iterrows():
    print(f"[{idx+1}/{len(relevant_datasets)}]", end=" ")
    try:
        df = download_socrata_full_safe(row)
        if df is not None:
            downloaded.append(row.name)
        else:
            failed.append(row.name)
    except:
        failed.append(row.name)
print(f"\nDONE!")
print(f"Successfully downloaded: {len(downloaded)}")
print(f"Failed/skipped: {len(failed)}")
if failed:
    print("Failed rows:", failed)
```
